```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('r2d2.jpg', 0)
plt.imshow(img, cmap='gray', vmin=0, vmax=255)
```

digital images are just 2D functions

```python
dark_r2d2 = np.where(img > 100, img - 100, 0)
plt.imshow(dark_r2d2, cmap='gray', vmin=0, vmax=255)
```

```python
light_r2d2 = np.where(img < 155, img + 100, 255)
plt.imshow(light_r2d2, cmap='gray', vmin=0, vmax=255)
```

```python
inverse_r2d2 = 255 - img
plt.imshow(inverse_r2d2, cmap='gray', vmin=0, vmax=255)
```

```python
log_r2d2 = np.ma.log(img).filled(0)
```

scale to full intensity range

```python
log_r2d2_scaled = log_r2d2 / (log_r2d2.max() - log_r2d2.min()) * 255
plt.imshow(log_r2d2_scaled, cmap='gray', vmin=0, vmax=255)
```

imshow scales automatically

```python
plt.imshow(log_r2d2, cmap='gray')
```

```python
gamma = 5.0
gamma_r2d2 = np.power(img, gamma)
plt.imshow(gamma_r2d2, cmap='gray')
```
